```python
# Necessary imports
import os
import warnings
warnings.filterwarnings('ignore')

# Changing os working directory to the dataset location   
location = "C:/Users/shree/Documents/IR Assignments/Assignment 1/CSE508_Winter2023_Dataset/CSE508_Winter2023_Dataset"
os.chdir(location)
```

```python
# A function for reading contents between <TITLE></TITLE> and <TEXT></TEXT> tags and writing it back to the same document
def reading_file(file_location,i):
    with open(file_location,'r+') as doc:
        # Read the file
        filetext = doc.read()
        
        title_open = "<TITLE>"
        title_close = "</TITLE>"
        text_open = "<TEXT>"
        text_close = "</TEXT>"
        
        # Set indices to opening and closing tags
        index1 = filetext.index(title_open)
        index2 = filetext.index(title_close)
        index3 = filetext.index(text_open)
        index4 = filetext.index(text_close)
        
        # Store contents between the tag <TITLE></TITLE> in string1 and <TEXT></TEXT> in string2
        string1 = filetext[index1+len(title_open)+1: index2]
        string2 = filetext[index3+len(text_open)+1: index4]
        
        # Prepare a replacement string
        replace_str=string1+" "+string2
        replace_str=replace_str.strip()
        
        # Print before processing contents for first 5 sample documents
        if(i<=5):
            print("\nOriginal Content of file "+os.path.basename(doc.name)+" is: \n")
            print(filetext)
            
        # Change file pointers to beginning of document and overwrite the document, and close it
        doc.seek(0)
        doc.truncate(0)
        doc.write(replace_str)
        doc.close()
        filetext = open(file_location,'r').read()
        
        # Print after processing contents for first 5 sample documents
        if(i<=5):
            print("\n\nContent of file "+os.path.basename(doc.name)+" After replacement is: \n")
            print(filetext)
            print('-----------------------------------------------------------------------------------------')
```

```python
# Index to keep printing before and after contents of first 5 files
i=1 

# Call the above function for each document in the corpus
for doc in os.listdir():
    file_location = f"{location}\{doc}"
    reading_file(file_location,i)
    i=i+1
```

### <b>3. Remove Stopwords</b>

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

# get english stopwords from nltk.corpus.stopwords 
stopwords_eng = stopwords.words('english')
```

```python
# Defininf a function for removing stopwords from passed list of tokens/words
def remove_stopwords(data):
    result=[]    
    for word in data:
        if word not in stopwords_eng:
            result.append(word)
    return result
   
i=0
stop_preprocess=[]
# Traverse through each document in the corpus 
for doc in os.listdir(location):
    # Print contents of first 5 documents before processing 
    if(i<5):
        print("\nContent of file "+doc+" before removing stopwords is: \n")
        print(token_preprocess[i])

    # Call the removing stopwords function remove_stopwords() on doc data
    doc_data = remove_stopwords(token_preprocess[i]) 
    stop_preprocess.append(doc_data)

    # Print after processing contents for first 5 sample documents
    if(i<5):
        print("\n\nContent of file "+doc+" after removing stopwords is: \n")
        print(stop_preprocess[i]) 
        print('-----------------------------------------------------------------------------------------')
    i=i+1
```

### <b>4. Remove Punctuations</b>

```python
# Necessary imports 
import string
import re

# Defining a function for removing punctuations 
def remove_punct(data):
    result = []
    result1 = []
    for word in data:
        if word not in string.punctuation:
            # Append the token to result list only if it is not present in predefined string.punctuation list
            result.append(word)
            
    # Additionally removing punctuations using regular expressions 
    result = [sub.replace('[^\w\s]',' ') for sub in result]
    for word in result:
        result1.append(re.sub(r'[^\w\s]','', word))
    return result1
   
i=0
pun_preprocess=[]
# Traverse through each document in the corpus 
for doc in os.listdir(location):
    # Print contents of first 5 documents before processing 
    if(i<5):
        print("\nContent of file "+doc+" before removing punctuations is: \n")
        print(stop_preprocess[i])

    # Call the punctuation removal function remove_punct() on doc data
    doc_data = remove_punct(stop_preprocess[i]);
    pun_preprocess.append(doc_data)

    # Print after processing contents for first 5 sample documents
    if(i<5):
        print("\n\nContent of file "+doc+" after removing punctuations is: \n")
        print(pun_preprocess[i]) 
        print('-----------------------------------------------------------------------------------------')
    i=i+1  
```

### <b>5. Remove blank space tokens</b>

```python
# Defining a functions for removing blank tokens
def remove_blank_tokens(data):
    result=[]
    
    # If the token is not blank, then append it to the result list
    for word in data:
        if word.strip()!='':
            result.append(word.strip())
    return result

        
i=0
blankspace_preprocess=[]
# Traverse through each document in the corpus 
for doc in os.listdir(location):    
    # Print contents of first 5 documents before processing 
    if(i<5):
        print("\nContent of file "+doc+" before removing blank tokens is: \n")
        print(pun_preprocess[i])

    # Call the blank tokens removal function remove_blank_tokens() on doc data
    doc_data = remove_blank_tokens(pun_preprocess[i])
    blankspace_preprocess.append(doc_data)
    
    # Print after processing contents for first 5 sample documents
    if(i<5):
        print("\n\nContent of file "+doc+" after removing blank tokens is: \n")
        print(blankspace_preprocess[i]) 
        print('-----------------------------------------------------------------------------------------')
    i=i+1

# Keeping record of final pre processed dataset 
final_dataset = blankspace_preprocess.copy()
```

## <b>3. Support for operations AND, OR, NOT</b>

```python
# Storing list of all documents in the corpus in all_docs
all_docs = list(os.listdir())

# A global variables to store minimum comparisons during query processing 
global total_comparisons
total_comparisons = 0
```

```python
# A fuction to get the set of all documents which does not contain the passed word
def get_not(word):
    global total_comparisons
    try:
        present_in = dictionary_postings[word]
    except:
        present_in = {}
    finally:
        all_docs = set(os.listdir())
        
    # During the NOT operations we traverse through all docs thus comaprisons required for NOT operation = total docs in corpus
    total_comparisons += len(all_docs)
    return all_docs.difference(present_in)     # Difference between all_docs and the set of docs which contain passed 'word'
```

```python
# A function to get the normal words and operators like 'and', 'or' in 2 separate lists 
def generate_operators_operands(query_tokens):
    #print("Initial query tokens generated : ",query_tokens)
    operators = []
    operands = []

    for word in query_tokens:
        if word not in ['and', 'or']:     # We process 'not' separately
            operands.append(word)         # Normal query term
        else:
            operators.append(word)        # Query operator 'and'/'or' 
    return operands, operators
```

```python
# A function to get the intersection of 2 sets 
def intersection(set1, set2):
    global total_comparisons
    result = []
    for item in set1:
        total_comparisons += 1            # total_comparisons increments by 1 for each iteration 
        if item in set2:
            result.append(item)
    return result

# A function to get the union of 2 sets
def union(set1, set2):
    result = []
    global total_comparisons
    for item in set1:
        total_comparisons += 1            # total_comparisons increments by 1 for each iteration
        if item not in result:
            result.append(item)
    for item in set2:
        total_comparisons += 1            # total_comparisons increments by 1 for each iteration
        if item not in result:
            result.append(item)
    return result
```

```python
# A function to process AND and OR given the list of query words ('operands') and operations like AND, OR, NOT 'operators' 
def and_or(operands, operators, not_list):
    # If operand contains integer then we take the corresponding NOT set from not_list[integer]
    if(type(operands[0])==int):
        a = not_list[operands[0]]
    else:
        # Process the initial operand and store resulting set in a
        try:
            # Get the term postings if that term is available in the unigram_inverted_index dictionary 
            a = dictionary_postings[operands[0]]
        except:
            #initialize set a to empty set {}, if term is not present in the unigram_inverted_index dictionary
            a = {}
        finally: 
            operands.pop(0)
            
    for i in range(len(operators)):
        # For each operand, take the inital set a and further intersect or unite with b 
        if type(operands[i])==int:              # If user gives a query like 'NOT term'
            b = not_list[operands[i]]
        else:
            try:
                b = dictionary_postings[operands[i]]
            except:
                b = {}
                
        # For each ith operator, take union/intersection between (i)th operand which is initially a and (i+1)th operand 
        if operators[i] == 'and':
            a = intersection(a,b)
        elif operators[i] == 'or':
            a = union(a,b)
        else:
            print("Invalid Operator")
    # Return final set after all operations
    return a
```

```python
def run_query(query):
    global total_comparisons

    # Clean the query 
    query = query.lower()
    query = query.strip()
    
    
    not_set_list = [] 
    tokens = query.split()
    
    # Prepare not_set for each 'not' occurrence in the general query 
    # Each part of query of type "NOT <term>" will be replaced with an integer stating the index of postings retrieved 
    # for that NOT <term> in the not_set_list
    not_count = 0
    for token in tokens:
        if token == 'not':
            not_count += 1
    for i in range(not_count):
        not_index = tokens.index('not')
        not_operand_index = not_index + 1
        
        not_tuple = get_not(tokens[not_operand_index])
        not_set_list.append(not_tuple)
        position = len(not_set_list)-1
        
        tokens.pop(not_index)
        tokens.pop(not_index)
        
        tokens.insert(not_index, int(position))
    
    # Generate operands (query words) and operators (operations) lists
    true_operands, true_operators = generate_operators_operands(tokens)   
    
    # Call and_or function for all operands an operators retrieved from the query 
    result = and_or(true_operands, true_operators, not_set_list)
        
    return result, total_comparisons
```

## <b>4, 5. Support for generalized boolean query (e.g. T1 AND T2 OR T3 OR NOT T4 ...)</b>

```python
# Input for generalized query
query = input("Enter your boolean query : ")

# Clean query 
query = query.lower()
query = query.strip()

# Initially set total_comparisions to 0
global total_comparisons 
total_comparisons = 0

# Run our main function 
result, total_comparisons = run_query(query)

result = list(result)
print("Number of documents retrieved for query : %d"%(len(result)))
print("Names of documents retrieved for query %s: "%(str()), end ="")
print(*result, sep=', ')
print("Number of comparisons required for query : %d"%(total_comparisons))
```
